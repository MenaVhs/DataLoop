### Chi cuadrada

```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv("Variables_entrevista.csv")

# Lista de columnas categóricas a codificar
columnas_categoricas = ['Modelo', 'EntidadFederativa', 'ActividadEconomica', 'Propiedad_Casa']
data_encoded = pd.get_dummies(df, columns=columnas_categoricas)
df = data_encoded.drop(['ID_Credito', 'FLG_Empleo', 'Monto_prestamo','IngresoMensual'], axis=1)
default_crediticio = df.pop('Default_Crediticio')
df['Default_Crediticio'] = default_crediticio # insertar Default_crediticio al final del df
df
```

```python
# Obtener todas las columnas excepto la última
columnas = df.columns[:-1]

for columna in columnas:
    tabla_contingencia = pd.crosstab(df[columna], df['Default_Crediticio'])
    
    # Realizar la prueba de chi-cuadrado
    chi2, p_valor, _, _ = chi2_contingency(tabla_contingencia)
    
    # Imprimir el p-valor
    print(f'El p-valor para la columna {columna} es: {p_valor}')
```

```python
import pandas as pd

columnas = df.columns[:-1]

resultados = []

for columna in columnas:
    tabla_contingencia = pd.crosstab(df[columna], df['Default_Crediticio'])
    
    chi2, p_valor, _, _ = chi2_contingency(tabla_contingencia)
    
    resultados.append(f'{columna}, {p_valor}')

# Guardar los resultados en un archivo de texto
with open('resultadosPorChiCuadrada.csv', 'w') as file:
    file.write('\n'.join(resultados))

print("Los resultados se han guardado en 'resultadosPorChiCuadrada.csv'")
```
